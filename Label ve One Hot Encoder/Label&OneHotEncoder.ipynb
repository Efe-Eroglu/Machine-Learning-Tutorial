```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("drive/MyDrive/Colab Notebooks/Veriler/veriler.csv")
```

```python
df.head()
```

```python
ulke = df.iloc[:,0:1].values
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
# LabelEncoder kullanarak ülke sütununu kodlar
ulke[:,0] =le.fit_transform(df.iloc[:,0])
```

```python
ulke
```

```python
# OneHotEncoder kullanarak ülke sütununu one-hot encode eder
ohe = preprocessing.OneHotEncoder()
ulke = ohe.fit_transform(ulke).toarray()
```

```python
ulke
```

```python
cinsiyet = df.iloc[:,-1:].values
```

```python
cinsiyet
```

```python
cinsiyet[:,-1] = le.fit_transform(df.iloc[:,-1])
```

```python
cinsiyet = ohe.fit_transform(cinsiyet).toarray()
```

```python
cinsiyet
```

```python
sonuc = pd.DataFrame(data=ulke, index=range(22), columns=["fr","tr","us"])
```

```python
sonuc
```

```python
sonuc2 = pd.DataFrame(data=cinsiyet[:,:1], index = range(22), columns= ["cinsiyet"])
```

```python
sonuc2
```

```python
sonuc3 = pd.DataFrame(data=df, index = range(22), columns = ['boy','kilo','yas'])
```

```python
sonuc3
```

```python
sonuc_df = pd.concat([sonuc,sonuc3],axis=1)
```

```python
sonuc_df2 = pd.concat([sonuc_df,sonuc2],axis=1)
```

```python
sonuc_df
```

```python
sonuc_df2
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Eğitim ve test kümelerine veri bölme işlemi yapar
x_train, x_test,y_train,y_test = train_test_split(sonuc_df,sonuc2,test_size=0.33, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
# Test kümesi üzerinde tahmin yapın
y_pred = regressor.predict(x_test)
```

```python
# Eğitim sonucu ile test sonucunu karşılaştırın
y_pred,y_test
```
