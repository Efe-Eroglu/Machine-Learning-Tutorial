```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Veriler/maaslar.csv")
```

```python
df
```

```python
egitim = df.iloc[:,1:2]  # Bağımsız değişken seçimi
maaslar = df.iloc[:,2:]  # Bağımlı değişken seçimi

#Numpy dizisi dönüşümü
EGITIM = egitim.values
MAASLAR = maaslar.values
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(EGITIM,MAASLAR)
```

```python
# Doğrusal olmayan model oluşturma (2.Dereceden)
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
egitim_poly =poly_reg.fit_transform(EGITIM)

lr2 = LinearRegression()
lr2.fit(egitim_poly, maaslar) #egitim_poly'ye göre maasları öğren
```

#Görselleştirne

```python
# Doğrusal regresyon sonuçları
plt.scatter(EGITIM,MAASLAR, color="red")
plt.plot(egitim,lr.predict(EGITIM), color="blue")
```

```python
# Polinom regresyon sonuçları
plt.scatter(EGITIM,MAASLAR, color="red")
plt.plot(EGITIM, lr2.predict(poly_reg.fit_transform(EGITIM)),color="blue")
```
