```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
df
```

```python
X = df.iloc[:, [2, 3]].values
y = df.iloc[:, 4].values
```

```python
X
```

```python
y
```

# Verileri Ayırma

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
```

# Normalizasyon

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# SVM

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score, KFold
kf = KFold(n_splits=5, shuffle=True, random_state=0)

# k-fold çapraz doğrulama yapın
scores = cross_val_score(classifier, X_train, y_train, cv=kf)

# Her kat için çapraz doğrulama skorlarını yazdırın
print("Her kat için çapraz doğrulama skorları:", scores)

# Çapraz doğrulama skorlarının ortalamasını ve standart sapmasını yazdırın
print("Ortalama çapraz doğrulama skoru:", np.mean(scores))
print("Çapraz doğrulama skorlarının standart sapması:", np.std(scores))
```

```python
y_pred = classifier.predict(X_test)
```

# Metrikler

```python
import seaborn as sns
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)

from sklearn.metrics import precision_score, recall_score, f1_score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)

conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Greens', cbar=False)
plt.xlabel('Tahminler')
plt.ylabel('True')
plt.title('Karmaşıklık Matrisi')
```
